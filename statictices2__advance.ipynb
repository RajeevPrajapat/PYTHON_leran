# Statistics Part 2



// 1. What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population based on sample data, typically by testing assumptions (hypotheses) about population parameters.

// 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis (H₀) is a statement that there is no effect or no difference, while the alternative hypothesis (H₁ or Ha) suggests that there is an effect or a difference.

// 3. What is the significance level in hypothesis testing, and why is it important?
- The significance level (α) is the probability of rejecting the null hypothesis when it is actually true; it defines the threshold for statistical significance, commonly set at 0.05.

// 4. What does a P-value represent in hypothesis testing?
- A P-value indicates the probability of observing the sample results, or more extreme, assuming that the null hypothesis is true.

// 5. How do you interpret the P-value in hypothesis testing?
- A small P-value (typically ≤ 0.05) indicates strong evidence against the null hypothesis, so it is rejected; a large P-value suggests weak evidence against the null, so it is not rejected.

// 6. What are Type 1 and Type 2 errors in hypothesis testing?
- Type I error is rejecting a true null hypothesis (false positive), and Type II error is failing to reject a false null hypothesis (false negative).

// 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- A one-tailed test checks for an effect in one direction only, while a two-tailed test checks for effects in both directions.

// 8. What is the Z-test, and when is it used in hypothesis testing?
- A Z-test is used when the population variance is known and the sample size is large; it determines if there is a significant difference between sample and population means.

// 9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
- Z = (X̄ - μ) / (σ / √n); it measures how many standard deviations a sample mean is from the population mean.

// 10. What is the T-distribution, and when should it be used instead of the normal distribution?
- The T-distribution is used when the sample size is small and the population standard deviation is unknown; it accounts for added uncertainty.

// 11. What is the difference between a Z-test and a T-test?
- Z-tests are used with known variance and large samples, while T-tests are used with unknown variance and smaller samples.

// 12. What is the T-test, and how is it used in hypothesis testing?
- A T-test assesses whether the means of two groups are statistically different, especially useful with small sample sizes.

// 13. What is the relationship between Z-test and T-test in hypothesis testing?
- Both test means but differ in assumptions: the Z-test assumes known variance and large samples, while the T-test is used for smaller samples or unknown variance.

// 14. What is a confidence interval, and how is it used to interpret statistical results?
- A confidence interval provides a range of values within which the true population parameter is expected to lie, with a certain level of confidence (e.g., 95%).

// 15. What is the margin of error, and how does it affect the confidence interval?
- The margin of error is the range around the sample estimate that is likely to contain the population parameter; a larger margin increases the confidence interval width.

// 16. How is Bayes’ Theorem used in statistics, and what is its significance?
- Bayes’ Theorem calculates the probability of a hypothesis given prior knowledge and new evidence, useful for updating beliefs based on data.

// 17. What is the Chi-square distribution, and when is it used?
- The Chi-square distribution is used in tests of independence and goodness of fit for categorical data.

// 18. What is the Chi-square goodness of fit test, and how is it applied?
- It checks how well observed categorical data match expected distributions; it's applied by comparing observed vs. expected frequencies.

// 19. What is the F-distribution, and when is it used in hypothesis testing?
- The F-distribution is used in analysis of variance (ANOVA) and tests comparing two variances.

// 20. What are the different types of ANOVA tests?
- Types include one-way ANOVA (one independent variable), two-way ANOVA (two independent variables), and repeated measures ANOVA (same subjects measured multiple times).

// 21. What is the F-test, and how does it relate to hypothesis testing?
- The F-test compares two variances to determine if they are significantly different; it's the basis for ANOVA tests.

// 22. What is an ANOVA test, and what are its assumptions?
- ANOVA (Analysis of Variance) is a statistical method used to compare the means of three or more groups to determine if at least one group mean is significantly different from the others. It helps identify whether observed differences among sample means are due to actual differences or random chance.

- The main assumptions of ANOVA are:
  1. **Independence of observations** – the data should be collected independently across groups.
  2. **Normality** – the residuals (errors) within each group should be approximately normally distributed.
  3. **Homogeneity of variances** – all groups should have similar variances (equal spread or dispersion).

```python

## 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
sample = np.random.normal(60, 10, 100)
pop_mean = 55
z_stat = (np.mean(sample) - pop_mean) / (np.std(sample, ddof=1) / np.sqrt(len(sample)))
p_val = 2 * (1 - stats.norm.cdf(abs(z_stat)))
print("Q3 - Z-test: Z =", z_stat, ", p =", p_val)
```

```python
## 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
x = np.linspace(-4, 4, 1000)
plt.plot(x, stats.norm.pdf(x))
plt.axvline(-1.96, color='red', linestyle='--')
plt.axvline(1.96, color='red', linestyle='--')
plt.title('Q4 - Two-tailed Z-test Critical Region')
plt.xlabel('Z-value')
plt.ylabel('Density')
plt.show()
```

```python
## 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
def visualize_type_errors(mu0=0, mu1=1, sigma=1, n=30, alpha=0.05):
    x = np.linspace(-4, 4, 1000)
    crit_val = stats.norm.ppf(1 - alpha, loc=mu0, scale=sigma/np.sqrt(n))
    plt.plot(x, stats.norm.pdf(x, loc=mu0, scale=sigma/np.sqrt(n)), label="H0")
    plt.plot(x, stats.norm.pdf(x, loc=mu1, scale=sigma/np.sqrt(n)), label="H1")
    plt.axvline(crit_val, color='red', linestyle='--', label="Critical Value")
    plt.fill_between(x, 0, stats.norm.pdf(x, loc=mu0, scale=sigma/np.sqrt(n)), where=(x > crit_val), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, 0, stats.norm.pdf(x, loc=mu1, scale=sigma/np.sqrt(n)), where=(x < crit_val), color='blue', alpha=0.3, label='Type II Error')
    plt.legend()
    plt.title("Q5 - Type I and Type II Errors")
    plt.show()

visualize_type_errors()
```

```python
## 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
# Example: P(A|B) = [P(B|A) * P(A)] / P(B)
P_A = 0.01
P_B_given_A = 0.9
P_B_given_not_A = 0.1
P_not_A = 1 - P_A
P_B = P_B_given_A * P_A + P_B_given_not_A * P_not_A
P_A_given_B = (P_B_given_A * P_A) / P_B
print("Q11 - Posterior Probability (Bayes' Theorem):", P_A_given_B)
```
