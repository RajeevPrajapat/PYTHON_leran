# 1. What is a random variable in probability theory?
("A random variable is a numerical value assigned to each possible outcome of a random experiment. "
      "It provides a way to quantify uncertain outcomes. There are two types: discrete and continuous. "
      "Random variables are fundamental in probability theory and statistics as they help in defining distributions, "
      "expected values, and performing statistical analysis based on uncertain events.")

# 2. What are the types of random variables?
("There are two main types of random variables: discrete and continuous. "
      "Discrete random variables take on a countable number of values, such as number of heads in coin tosses. "
      "Continuous random variables take on any value within a given range, like height or temperature. "
      "These types help in determining appropriate statistical methods and probability distributions.")

# 3. What is the difference between discrete and continuous distributions?
("Discrete distributions describe variables with specific, separate values, such as the binomial distribution. "
      "Continuous distributions describe variables that can take any value within a range, such as the normal distribution. "
      "In discrete distributions, probabilities are assigned to individual outcomes, whereas in continuous ones, "
      "they are described using a probability density function (PDF).")

# 4. What are probability distribution functions (PDF)?
("A Probability Distribution Function (PDF) defines the probability of a continuous random variable falling "
      "within a particular range. It does not provide probabilities for exact values but instead indicates density. "
      "The area under the curve of the PDF between two points represents the probability of the variable lying in that range. "
      "PDFs are essential in understanding distributions like normal and exponential.")

# 5. How do cumulative distribution functions (CDF) differ from PDF?
("A Cumulative Distribution Function (CDF) provides the probability that a random variable is less than or equal to a value. "
      "In contrast, a Probability Density Function (PDF) represents the probability density at a point. "
      "The CDF is obtained by integrating the PDF. While PDF indicates likelihood at an instant, CDF gives accumulated probability.")

# 6. What is a discrete uniform distribution?
("A discrete uniform distribution is a probability distribution where each value within a finite set of outcomes has equal probability. "
      "An example is a fair six-sided die, where each face (1 to 6) has a probability of 1/6. "
      "This distribution is simple yet useful in scenarios where outcomes are equally likely.")

# 7. What are the key properties of a Bernoulli distribution?
("The Bernoulli distribution is a discrete distribution with only two possible outcomes: success (1) and failure (0). "
      "It is characterized by a single parameter 'p', the probability of success. "
      "Its mean is 'p' and variance is 'p(1 - p)'. It forms the basis for the binomial distribution, which is the sum of multiple Bernoulli trials.")

# 8. What is the binomial distribution, and how is it used in probability?
("The binomial distribution describes the number of successes in a fixed number of independent Bernoulli trials, each with probability 'p'. "
      "It is used when outcomes are binary (success/failure) and is widely applied in quality control, voting surveys, and medical testing. "
      "The mean is 'np' and variance is 'np(1-p)'. It models discrete random variables effectively.")

# 9. What is the Poisson distribution and where is it applied?
("The Poisson distribution models the number of times an event occurs in a fixed interval of time or space, "
      "given a known constant mean rate and independent occurrences. "
      "It is used in fields such as telecommunications (e.g., call arrivals), traffic flow, and biology. "
      "The mean and variance are both equal to the parameter λ (lambda), the average rate.")

# 10. What is a continuous uniform distribution?
("A continuous uniform distribution describes a scenario where every outcome in a continuous range is equally likely. "
      "For example, selecting a random number between 0 and 1. "
      "The PDF is constant across the range, and the mean is the average of the lower and upper bounds. "
      "It is a simple model for evenly distributed continuous data.")

# 11. What are the characteristics of a normal distribution?
("A normal distribution, also called a Gaussian distribution, is symmetric and bell-shaped. "
      "Its mean, median, and mode are all equal. It is characterized by two parameters: mean (μ) and standard deviation (σ). "
      "Many natural phenomena, like heights or test scores, follow a normal distribution. "
      "It plays a key role in statistical inference due to the Central Limit Theorem.")

# 12. What is the standard normal distribution, and why is it important?
("The standard normal distribution is a special case of the normal distribution with a mean of 0 and a standard deviation of 1. "
      "It allows for comparison of scores from different distributions using Z-scores. "
      "Statistical tables and tools often rely on the standard normal distribution to compute probabilities and critical values.")

# 13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
("The Central Limit Theorem (CLT) states that the distribution of sample means approaches a normal distribution as the sample size increases, "
      "regardless of the original population's distribution. "
      "It is crucial in statistics because it justifies the use of normal approximation in hypothesis testing and confidence intervals, even for non-normal populations.")

# 14. How does the Central Limit Theorem relate to the normal distribution?
("The Central Limit Theorem explains why the normal distribution appears so frequently in statistics. "
      "As sample size increases, the sampling distribution of the sample mean tends to become normal, even if the data itself is not. "
      "This enables the use of Z-scores, confidence intervals, and hypothesis tests for large samples.")

# 15. What is the application of Z statistics in hypothesis testing?
("Z statistics are used to determine whether to reject a null hypothesis by comparing a calculated Z-value to a critical value. "
      "This approach assumes the population standard deviation is known. "
      "It’s commonly used in comparing sample and population means, especially when the sample size is large or the population is normally distributed.")

# 16. How do you calculate a z-score, and what does it represent?
("The z-score is calculated using the formula z = (x - μ) / σ, where x is a data point, μ is the mean, and σ is the standard deviation. "
      "It measures how far a value is from the mean in terms of standard deviations. "
      "Z-scores are helpful for detecting outliers and standardizing data from different distributions.")

# 17. What are point estimates and interval estimates in statistics?
("A point estimate provides a single best guess of a population parameter, such as a sample mean estimating the population mean. "
      "An interval estimate gives a range (confidence interval) within which the parameter is likely to lie. "
      "Point estimates are precise but less informative, while interval estimates offer a range of values with confidence levels.")

# 18. What is the significance of confidence intervals in statistical analysis?
("Confidence intervals represent the range within which we expect a population parameter to lie, with a certain confidence level (e.g., 95%). "
      "They provide more information than point estimates by incorporating the uncertainty in sampling. "
      "They are widely used in research, surveys, and decision-making to reflect the precision of an estimate.")

# 19. What is the relationship between a Z-score and a confidence interval?
("Z-scores define the boundaries of confidence intervals in standard normal distributions. "
      "For example, a 95% confidence interval corresponds to a Z-score of ±1.96. "
      "By multiplying the standard error by the Z-score, we get the margin of error, which helps construct the interval around the estimate.")

# 20. How are Z-scores used to compare different distributions?
("Z-scores standardize values by converting them into the number of standard deviations from the mean. "
      "This allows comparison across different distributions or datasets, regardless of the original scale. "
      "They are essential in detecting outliers and conducting standard normal hypothesis testing.")

# 21. What are the assumptions for applying the Central Limit Theorem?
("To apply the Central Limit Theorem, the sample observations should be independent, identically distributed, and the sample size should be sufficiently large. "
      "Typically, a sample size greater than 30 is considered adequate. "
      "The CLT holds for many distributions, which makes it a powerful tool in statistics.")

# 22. What is the concept of expected value in a probability distribution?
("The expected value is the mean or average of all possible outcomes of a random variable, weighted by their probabilities. "
      "It represents the long-run average result of a random experiment. "
      "For discrete variables, it is calculated as Σ[x * P(x)]. It is a fundamental concept in decision theory and statistics.")
